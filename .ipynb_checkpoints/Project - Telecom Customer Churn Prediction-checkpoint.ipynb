# Telecom Customer Churn Prediction

```python
# import packages

#Data Structures
import pandas as pd
import numpy as np
import re
from IPython.display import display

#Sklearn
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix, precision_score, recall_score
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold

#Plotting
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# self_defined
import functions
```

## I. Read and Preprocess Data

### I.1 read data

```python
# Read data
data_df = pd.read_csv("data/telecom-churn-case-study-hackathon-c52/train.csv")
data_df.shape
```

```python
data_df.describe()
```

### I.2 split training and testing

```python
X = data_df.iloc[:,:-1]
y = data_df.iloc[:,-1]

# split train and test
Xtrain_df, Xtest_df, ytrain_df, ytest_df = train_test_split(X, y,
                                                            #random_state = 99,
                                                            test_size = 0.2)
```

```python
Xtrain_df.shape, Xtest_df.shape, ytrain_df.shape, ytest_df.shape
```

### I.2 exclude meaningless columns in training data


```python
print("Original training data size: ", Xtrain_df.shape)
```

```python
# drop "id" column
Xtrain_df = Xtrain_df.drop("id", axis = 1)
```

```python
# Excldue columns showing dates
Xtrain_df = functions.exclude_date_fields(Xtrain_df)
```

```python
# exclude columns with no variations
Xtrain_df = functions.exclude_single_value_fields(Xtrain_df)
```

```python
# Exclude columns with too many missng values
Xtrain_df =  functions.exclude_fields_missing_values(Xtrain_df, 0.2)
```

```python
print("Processed training data size: ", Xtrain_df.shape)

# store the columns needed in the following steps into a list
columns_needed = Xtrain_df.columns

print("Columns remaining in the dataset: ", len(columns_needed))
```

### I.3 impute missing values in training data

```python
# show the percentage of null values in each column
functions.show_missing_value_percent(Xtrain_df, "Percentage of missing values for each column before imputation")
Xtrain_df.describe()
```

```python
imputation = SimpleImputer(strategy='mean').fit(Xtrain_df)

Xtrain_df[columns_needed] = imputation.transform(Xtrain_df)
```

```python
functions.show_missing_value_percent(Xtrain_df, "Percentage of missing values for each column after imputation")
Xtrain_df.describe()
```

### I.4 Handle outlier

```python
#restrict outliers of each field to be within k = 3 times of standardard deviation from mean

#Xtrain_df = Xtrain_df.apply(functions.cap_outliers, axis = 0)
#Xtrain_df.describe().round()
```

### I.5 standardize the data

```python
standardization = StandardScaler().fit(Xtrain_df)

Xtrain_df[columns_needed] = standardization.transform(Xtrain_df)
```

```python
Xtrain_df.describe().round()
```

```python
# convert data into numpy array
Xtrain = np.array(Xtrain_df)
ytrain = np.array(ytrain_df)

Xtrain.shape, ytrain.shape
```

### I.6 PCA

```python
# Correlation heat map of original variables
plt.figure(figsize=(10,8))
corr_heatmap = sns.heatmap(Xtrain_df.corr())
corr_heatmap.set_title("Correlation heatmap of original variables")
plt.tight_layout()
plt.savefig('plot\Correlation heatmap of original variables.png')
plt.show()
```

```python
pca = PCA(n_components = 70).fit(Xtrain)

Xtrain_pca = pca.transform(Xtrain)
display(pd.DataFrame(Xtrain_pca))

print("Explained Variance Ratio after PCA: ", np.sum(pca.explained_variance_ratio_))
```

```python
# show a few scatter plots made of PCA components
functions.PCA_scatter(4, Xtrain_pca, ytrain)
```

```python
# Correlation heat map of PCA components
plt.figure(figsize=(8, 6))
corr_heatmap = sns.heatmap(np.corrcoef(Xtrain_pca.T))
corr_heatmap.set_title("Correlation heatmap of PCA variables")
plt.tight_layout()
plt.savefig('plot\Correlation heatmap of PCA variables.png')
plt.show()
```

### I.7 Process Testing Data

```python
# process test data
Xtest_df = Xtest_df[columns_needed]
Xtest_df[columns_needed] = imputation.transform(Xtest_df)
Xtest_df[columns_needed] = standardization.transform(Xtest_df)

Xtest = np.array(Xtest_df)
ytest = np.array(ytest_df)

Xtest_pca = pca.transform(Xtest)
```

### I.8 create indices for 5 fold cross validation

```python
cv = KFold(n_splits = 5, shuffle = True, random_state = 100)

for i, (train_index, test_index) in enumerate(cv.split(Xtrain)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}     ", len(train_index))
    print(f"  Test:  index={test_index}     ", len(test_index))
```

## III. Cross Validation

### III.1 group classifier and classifier names

```python
results_df = []
```

#### III.1.1 Logistic Regression & KNN

```python
classifiers = [LogisticRegression(max_iter = 10000, 
                                  class_weight = 'balanced',
                                  penalty = None,
                                  solver = 'lbfgs'),
               KNeighborsClassifier(n_neighbors = 5,
                                    weights = "distance",
                                    p = 2),
               KNeighborsClassifier(n_neighbors = 6,
                                    weights = "distance",
                                    p = 2),
               KNeighborsClassifier(n_neighbors = 7,
                                    weights = "distance",
                                    p = 2),
               KNeighborsClassifier(n_neighbors = 8,
                                    weights = "distance",
                                    p = 2),
               KNeighborsClassifier(n_neighbors = 9,
                                    weights = "distance",
                                    p = 2),
               KNeighborsClassifier(n_neighbors = 10,
                                    weights = "distance",
                                    p = 2)]

classifier_names = ['Logistic Regression - PCA variables',
                    'KNN - PCA variables, k = 5',
                    'KNN - PCA variables, k = 6',
                    'KNN - PCA variables, k = 7',
                    'KNN - PCA variables, k = 8',
                    'KNN - PCA variables, k = 9',
                    'KNN - PCA variables, k = 10']
```

```python
for classifier, classifier_name in zip(classifiers, classifier_names):
    result = functions.cross_validation(classifier, 
                                        classifier_name = classifier_name,
                                        Xtrain = Xtrain_pca, 
                                        ytrain = ytrain,
                                        threshold = 0.5,
                                        cv = cv) 

    #display(result)
    if len(results_df) == 0:
        results_df = [result]
    else:
        results_df += [result]
```

#### III.1.2 Neural Network

```python
classifiers = [MLPClassifier(hidden_layer_sizes = (200, 100, 50, 30, 30, 30),
                             max_iter = 1000,
                             solver = "adam",
                             learning_rate = "adaptive",
                             learning_rate_init = 0.001,
                             activation = "relu",
                             alpha = 0.0001,
                             early_stopping = True,
                             tol = 1e-20,
                             batch_size = 100,
                             n_iter_no_change = 10)]

classifier_names = ["Neural Network - original variables"]
```

```python
for classifier, classifier_name in zip(classifiers, classifier_names): 
    result = functions.cross_validation(classifier, 
                                        classifier_name = classifier_name,
                                        Xtrain = Xtrain, 
                                        ytrain = ytrain,
                                        threshold = 0.5,
                                        cv = cv) 
    #display(result)
    if len(results_df) == 0:
        results_df = [result]
    else:
        results_df += [result]
```

#### III.1.3 Random Forest & AdaBoost

```python
classifiers = [RandomForestClassifier(n_estimators = 50),
               RandomForestClassifier(n_estimators = 100),
               RandomForestClassifier(n_estimators = 150),
               RandomForestClassifier(n_estimators = 200),
               AdaBoostClassifier(estimator = DecisionTreeClassifier(max_depth = 1),
                                  n_estimators = 50),
               AdaBoostClassifier(estimator = DecisionTreeClassifier(max_depth = 1),
                                  n_estimators = 100),
               AdaBoostClassifier(estimator = DecisionTreeClassifier(max_depth = 1),
                                  n_estimators = 150),
               AdaBoostClassifier(estimator = DecisionTreeClassifier(max_depth = 1),
                                  n_estimators = 200)]

classifier_names = ['Random Forest - original variables, n = 50', 
                    'Random Forest - original variables, n = 100',
                    'Random Forest - original variables, n = 150',
                    'Random Forest - original variables, n = 200',
                    "AdaBoost - decision stump with original variables, n = 50",
                    "AdaBoost - decision stump with original variables, n = 100",
                    "AdaBoost - decision stump with original variables, n = 150",
                    "AdaBoost - decision stump with original variables, n = 200"]
```

```python
for classifier, classifier_name in zip(classifiers, classifier_names): 
    result = functions.cross_validation(classifier, 
                                        classifier_name = classifier_name,
                                        Xtrain = Xtrain, 
                                        ytrain = ytrain,
                                        threshold = 0.5,
                                        cv = cv) 
    #display(result)
    if len(results_df) == 0:
        results_df = [result]
    else:
        results_df += [result]
```
